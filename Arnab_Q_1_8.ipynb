```r
library(igraph)
```

```r
g=read_graph("facebook_combined.txt",directed="False")
```

```r
plot(g,vertex.size=4, vertex.label.cex=0.1,vertex.color='blue')
```

## Q1

```r
is.connected(g)
```

## Q2

```r
print(sprintf("The vertex count is %s",vcount(g)))
print(sprintf("The edge count is %s",ecount(g)))
```

```r
print(sprintf("The diameter of the network is %s",diameter(g)))
```

## Q3

```r
plot(degree.distribution(g),type='l',main = "Degree Distribution", xlab = "Degree", ylab = "Frequency")
```

```r
print(sprintf("The slope of the line is %s",mean(degree(g))))
```

## Q4

```r
plot(degree.distribution(g),log = "xy",main = "Degree Distribution (log-log)", xlab = "Degree", ylab = "Frequency")
line = lm(log10(degree.distribution(g)) ~ log10(1:length(degree.distribution(g))), subset = degree.distribution(g) > 0)
abline(line, col = "red")
```

```r
print(sprintf("The slope of the line is %s",coef(line)[2]))
```

## Q5

```r
g_node1 = induced.subgraph(g, c(1, neighbors(g,1)))
```

```r
print(sprintf("The vertex count is %s",vcount(g_node1)))
print(sprintf("The edge count is %s",ecount(g_node1)))
```

## Q8

```r
count=0
sum=0
for (d in degree(g)) 
{
    if (d > 200) 
    {
        count=count + 1
        sum=sum + d
    }
}
print(sprintf("There are %s core nodes",count))
print(sprintf("Their sum is %s",sum))
print(sprintf("Their average is %s",(sum/count)))
```
